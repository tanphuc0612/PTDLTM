```
import pandas as pd
import numpy as np
```

```
mnist = pd.read_csv('https://raw.githubusercontent.com/tanphuc0612/PTDLTM/master/data/train.csv')
mnist.head(5)
```

```
data_test = pd.read_csv('https://raw.githubusercontent.com/tanphuc0612/PTDLTM/master/data/test.csv')
data_test.head(5)
```

```
train = mnist.values[:,1:].astype(float)
train_labels = mnist.values[:,0]
test = data_test.values[:,1:].astype(float)
test_labels = data_test.values[:,0]
```

```
import matplotlib.pyplot as plt

for i in range(10):
  image = train[i].reshape((28,28))
  label = train_labels[i]

  plt.figure(figsize = (15,2))
  imgplot = plt.imshow(image,cmap=plt.cm.binary)
  plt.show()
  print("Label:",label)
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit on training set only.
scaler.fit(train)

# Apply transform to both the training set and the test set.
train_normalized_img = scaler.transform(train)
test_normalized_img = scaler.transform(test)
```

```
# Model with all features
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
 
lm=linear_model.LogisticRegression(fit_intercept=True, solver = "sag", max_iter=100)
lm.fit(train_normalized_img,train_labels)
lm.score(test_normalized_img,test_labels)
```
